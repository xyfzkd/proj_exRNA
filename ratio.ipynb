```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
proj_exRNA_read = '/home/xieyufeng/proj_exRNA/stat/proj_exRNA.readsN.stat.txt'
```

```python
read_type = pd.read_table(proj_exRNA_read)
print(read_type)
```

```python
read_type['rRNA_ratio'] = read_type['rRNAN']/read_type['cleanN']
read_type['miRNA_ratio'] = read_type['miRNAN']/read_type['cleanN']
read_type['piRNA_ratio'] = read_type['piRNAN']/read_type['cleanN']
read_type['Y_RNA_ratio'] = read_type['Y_RNAN']/read_type['cleanN']
read_type['srpRNA_ratio'] = read_type['srpRNAN']/read_type['cleanN']
read_type['tRNA_ratio'] = read_type['tRNAN']/read_type['cleanN']
read_type['snRNA_ratio'] = read_type['snRNAN']/read_type['cleanN']
read_type['snoRNA_ratio'] = read_type['snoRNAN']/read_type['cleanN']
read_type['lncRNA_ratio'] = read_type['lncRNAN']/read_type['cleanN']
read_type['mRNA_ratio'] = read_type['mRNAN']/read_type['cleanN']
read_type['tucp_ratio'] = read_type['tucpN']/read_type['cleanN']
read_type['hg38other_ratio'] = read_type['hg38otherN']/read_type['cleanN']
read_type['other_ratio'] = read_type['de_hg38otherN']/read_type['cleanN']
rnanames = ['rRNA','miRNA','piRNA','Y_RNA','srpRNA','snRNA','snoRNA','lncRNA','mRNA','tucp','hg38other','other']
read_type
```

```python
read_type_info = read_type[['rRNA_ratio','miRNA_ratio','piRNA_ratio','Y_RNA_ratio','srpRNA_ratio','tRNA_ratio','snRNA_ratio','snoRNA_ratio','lncRNA_ratio','mRNA_ratio','tucp_ratio','hg38other_ratio','other_ratio']]
read_type_info['sum']=read_type_info.apply(lambda x: x.sum(), axis=1)
read_type_info
```
